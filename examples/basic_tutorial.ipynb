```python
import librosa
from librosa import display, load
import matplotlib.pyplot as plt
import numpy as np
import torch
from torch import nn
import torch.nn.functional as F
from torchnmf import *
from scipy.signal import fftconvolve
```

# Basic


## Define Target Signal


```python
y, sr = librosa.load(librosa.util.example_audio_file())
y = torch.from_numpy(y)
windowsize = 2048
S = torch.stft(y, windowsize, window=torch.hann_window(windowsize)).pow(2).sum(2).sqrt()
S = torch.FloatTensor(S.T)
S[S == 0] = 1e-8
```

## NMF

```python
model1 = NMF(S.shape, rank=10).cuda()
print(model1)
model1.sparse_fit(S.cuda(),
                  verbose=True, beta=1, max_iter=500,
                  sW=0.7)
```
